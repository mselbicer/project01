```python
df2 = pd.read_csv('airlines.csv')
df2
```

```python
iatacodes = pd.merge(df,df2, left_on="AIRLINE", right_on='IATA_CODE')
iatacodes
```

```python
iatacodes.rename(columns={'AIRLINE_y': 'AIRLINE_NAME', 'AIRLINE_x': 'IATA_CODE'}, inplace=True)
iatacodes
```

```python
# create a list of our conditions
cond_climates = [
    (iatacodes['MONTH'] < 3 ) | (iatacodes['MONTH'] == 12),
    (iatacodes['MONTH'] >= 3) & (iatacodes['MONTH'] < 6),
    (iatacodes['MONTH'] >= 6) & (iatacodes['MONTH'] < 9),
    (iatacodes['MONTH'] >= 9) & (iatacodes['MONTH'] < 12)
    ]

# create a list of the values we want to assign for each condition
values_climate = ['Winter', 'Spring', 'Summer', 'Autumn']

# create a new column and use np.select to assign values to it using our lists as arguments
iatacodes['CLIMATE'] = np.select(cond_climates, values_climate)

# display updated DataFrame
iatacodes.sample(7)
```

```python
flight_no = iatacodes.AIRLINE_NAME.value_counts()
flight_no
```
